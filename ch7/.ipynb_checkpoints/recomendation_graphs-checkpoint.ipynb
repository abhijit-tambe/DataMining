```python
import spacy
from sklearn.base import TransformerMixin
nlp = spacy.load('en')
class BagOfWords(TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        results = []
        for document in X:
            row = {}
            for word in list(nlp(document)):
                if len(word.text.strip()):
                    row[word.text] = True
                    results.append(row)
        return results
```
